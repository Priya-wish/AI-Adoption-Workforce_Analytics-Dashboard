## Exploratory Data Analysis

## Salaries Distribution Analysis

```python
plt.figure(figsize=(8,5))
plt.hist(df["Salary_USD"], bins=10, edgecolor="black")
plt.title("Distribution of AI Job Salaries"===
plt.xlabel("Salary (USD)")
plt.ylabel("Number of Jobs")
plt.show()
```

## AI Adoption Level Analysis

```python
df["AI_Adoption_Level"].value_counts().plot(
    kind="bar",
    color=["green","orange","red"]
)

plt.title("AI Adoption Level")
plt.xlabel("Adoption Level")
plt.ylabel("Count")
plt.show()
```

## Automation Risk Analysis

```python
df["Automation_Risk"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Automation Risk Distribution")
plt.ylabel("")
plt.show()
```

## Required Skills Analysis

```python
df["Required_Skills"].value_counts().head(10).plot(
    kind="barh",
    color="skyblue"
)

plt.title("Top Required AI Skills")
plt.xlabel("Count")
plt.ylabel("Skills")
plt.show()
```

## Salary Distribution by Company size

```python
plt.figure(figsize=(8,5))

df.boxplot(column="Salary_USD", by="Company_Size")

plt.title("Salary Distribution by Company Size")
plt.suptitle("")
plt.xlabel("Company Size")
plt.ylabel("Salary (USD)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```

# Business Insights

1. Most AI job salaries are concentrated between **USD 80,000 and USD 110,000**, indicating a competitive salary range.

2. **Medium AI adoption** is the most common adoption level among organizations, followed closely by Low adoption.

3. Automation risk is almost evenly distributed across Low, Medium, and High categories, suggesting that AI affects different job roles differently.

4. Skills such as **Project Management, Python, Cybersecurity, Machine Learning, Data Analysis, and UX/UI Design** appear among the most frequently required skills.

5. Salary distributions are fairly similar across Small, Medium, and Large companies, although a few high-paying outliers exist.

# Conclusion
